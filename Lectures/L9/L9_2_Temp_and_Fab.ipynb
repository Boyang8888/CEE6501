## Part 1 — Temperature Changes Conceptual Framework

### Why Temperature Changes Matter

Like support settlements, temperature changes can induce significant stresses in a structure.

If a member were completely free, it would simply:

- expand or contract axially, and/or
- bend due to a temperature gradient through its depth.

But in a statically indeterminate structure, surrounding members and supports restrain these free deformations.

That restraint produces **member fixed-end forces**.


### Connection to Support Settlement

The mechanics are very similar to support settlement:

- A member has some **free deformation** it would like to undergo
- The structure restrains that deformation
- Restraining it creates a **fixed-end force vector**

The main difference is the coordinate system in which the effect is most naturally described:

- **Support settlement** is usually prescribed in the **global** system
- **Temperature changes** are usually defined in the **local member** system
- **Fabrication errors** are also usually defined in the **local member** system


### Core Idea

We first determine the **free local end displacements** caused by the temperature change.

Then we ask:

> What local end forces would be required to suppress those displacements?

That force vector is the local thermal fixed-end force vector:

$$
\mathbf{q}^{H}
$$

This is exactly the same superposition logic used earlier for member loads and support settlement.




### Notation

$H$ superscript indicates FEFs due to temperature change.

Assume the start node is $b$ and the end node is $e$.

$$
\mathbf{q}^{H} =
\begin{bmatrix}
q_1^{H} \\[1pt]
q_2^{H} \\[1pt]
q_3^{H} \\[1pt]
q_4^{H} \\[1pt]
q_5^{H} \\[1pt]
q_6^{H}
\end{bmatrix}
=
\begin{bmatrix}
q_{xb}^{H} \\[1pt]
q_{yb}^{H} \\[1pt]
q_{mb}^{H} \\[1pt]
q_{xe}^{H} \\[1pt]
q_{ye}^{H} \\[1pt]
q_{me}^{H}
\end{bmatrix}
=
\begin{bmatrix}
N_b^{H} \\[1pt]
V_b^{H} \\[1pt]
M_b^{H} \\[1pt]
N_e^{H} \\[1pt]
V_e^{H} \\[1pt]
M_e^{H}
\end{bmatrix}
$$


We write local fixed-end forces using the convention:

- $q_{xb}^H$ = FEF at node **b** in local **x** (axial) → corresponds to **DOF 1**
- $q_{yb}^H$ = FEF at node **b** in local **y** (shear) → corresponds to **DOF 2**
- $q_{mb}^H$ = FEF at node **b** (moment) → corresponds to **DOF 3**

- $q_{xe}^H$ = FEF at node **e** in local **x** (axial) → corresponds to **DOF 4**
- $q_{ye}^H$ = FEF at node **e** in local **y** (shear) → corresponds to **DOF 5**
- $q_{me}^H$ = FEF at node **e** (moment) → corresponds to **DOF 6**

So the fixed-end force vector is
$$
\mathbf{q}^{H,T}
=
\begin{bmatrix}
q_{xb}^{H} & q_{yb}^{H} & q_{mb}^{H} &
q_{xe}^{H} & q_{ye}^{H} & q_{me}^{H}
\end{bmatrix}^T
$$


## Part 2 — Temperature Changes: Simply Supported Member

### Temperature Field Assumption

Consider a plane-frame member of length $L$ and depth $d$.

Assume:

- top surface temperature increase = $T_t$
- bottom surface temperature increase = $T_b$
- temperature varies **linearly through the depth**
- temperature does **not** vary along the member length
- coefficient of thermal expansion = $\alpha$

We assume the member is initially **simply supported**, so it is free to deform without developing end forces.


### Temperature Change over Member Depth

  <img src="assets/L2_Temp_Gradient.png" style="width:88%;">
</div>


### Axial Elongation of the Centroidal Axis

The temperature increase at the centroidal axis is the average of the top and bottom values:

$$
\frac{T_t + T_b}{2}
$$

So the axial elongation of the member is

$$
\Delta_H = \alpha \left( \frac{T_t + T_b}{2} \right)L
$$

where $\alpha$ is the coefficient of thermal expansion, with units of strain per degree of temperature change (for example, $/^\circ\mathrm{C}$).

This quantity tells us how much a material wants to expand or contract per unit length for each degree of temperature change.

Thus, $\Delta_{H}$ is the **free axial expansion** caused by the **average temperature increase** across the member depth.



Typical values of the coefficient of thermal expansion:

- **Concrete**: $ \alpha \approx (8\text{–}12)\times10^{-6} \; /^\circ\mathrm{C} $
- **Structural steel**: $ \alpha \approx 12\times10^{-6} \; /^\circ\mathrm{C} $
- **Wood (Parallel to grain)**: $ \alpha \approx (3\text{–}5)\times10^{-6} \; /^\circ\mathrm{C} $  
- **Wood (Perpendicular to grain)**: $ \alpha \approx (30\text{–}60)\times10^{-6} \; /^\circ\mathrm{C} $
- **Aluminum** $ \alpha \approx (22\text{–}24)\times10^{-6} \; /^\circ\mathrm{C} $

### End Rotations Due to Temperature Gradient

If $T_b \ne T_t$, the top and bottom fibers elongate by different amounts.

That causes the member to curve.

The magnitude of the end rotation (assuming $\theta$ is small) is

$$
\theta_H \approx \frac{\Delta_b - \Delta_t}{d} = \frac{\alpha (T_b - T_t)L}{2d}
$$

Using our sign convention:

- rotation at the **begin node** $b$ is clockwise, so it is **negative**
- rotation at the **end node** $e$ is counterclockwise, so it is **positive**


<div style="text-align:center;">
  <img src="assets/L2_Temp_FreeDisp.png" style="width:99%;">
</div>


### Local End Displacement Vector Due to Temperature

Using the local frame element DOF ordering

$$
\mathbf{v}^{T}
=
\begin{bmatrix}
v_1 & v_2 & v_3 & v_4 & v_5 & v_6
\end{bmatrix}^T
=
\begin{bmatrix}
v_{xb} & v_{yb} & \theta_b & v_{xe} & v_{ye} & \theta_e
\end{bmatrix}^T
$$

the free local end displacement vector, for a simply supported beam, due to temperature change is

$$
\mathbf{v}^{H}
=
\begin{bmatrix}
0 \\[4pt]
0 \\[4pt]
-\theta_H \\[4pt]
\Delta_H \\[4pt]
0 \\[4pt]
\theta_H
\end{bmatrix}
$$

where $v_3$ is negative because the rotation is clockwise at the starting node


substituting the expressions for $\Delta_H$ and $\theta_H$,

$$
\mathbf{v}^{H}
=
\frac{\alpha L}{2}
\begin{bmatrix}
0 \\[4pt]
0 \\[4pt]
-(T_b - T_t)/d \\[4pt]
T_b + T_t \\[4pt]
0 \\[4pt]
(T_b - T_t)/d
\end{bmatrix}
$$


### Reading the Temperature Displacement Vector

$$
\mathbf{v}^{H}
=
\begin{bmatrix}
0 & 0 & -\theta_T & \Delta_T & 0 & \theta_T
\end{bmatrix}^{T}
$$

Interpretation:

- no transverse joint translations: $v_2 = v_5 = 0$
- no axial motion at the start node: $v_1 = 0$
- axial expansion appears at the end node: $v_4 = \Delta_T$
- equal and opposite end rotations appear due to curvature: $v_3 = -\theta_T$ and $v_5 = \theta_T$

This is a **free deformation vector** for the simply supported member.


## Part 3 — Temperature Changes: Equivalent Fixed-End Forces

### Superposition Process

#### (1) Fixed–Fixed Member: Temperature Change
A prismatic frame member is restrained against end displacements and rotations.  
A uniform temperature increase induces **thermal strain**, which is prevented by the supports, generating **fixed-end forces**.

<div style="text-align:center;">
  <img src="assets/L2_Temp_FEF1.png" style="width:85%;">
</div>


#### (2) Simply Supported Member: Temperature Change Only
The same member is now simply supported and subjected to the same temperature increase.

- Ends are free to expand
- **No end forces develop**
- Thermal expansion produces end displacements, $v^{H}$

<div style="text-align:center;">
  <img src="assets/L2_Temp_FEF2.png" style="width:85%;">
</div>




#### (3) Simply Supported Member: End Forces Only
Apply the **fixed-end forces** from Case (1), but with **no temperature change**.

- Mechanical loading only
- Produces end displacements opposite to thermal expansion

<div style="text-align:center;">
  <img src="assets/L2_Temp_FEF3.png" style="width:85%;">
</div>



### Superposition Principle

The restrained (fixed–fixed) member response equals the sum of two simply supported cases:

$$
\text{Fixed member}
=
\text{Temperature-only}
+
\text{End-force-only}
$$

Displacement form:

$$
\mathbf{v}^{(1)} = \mathbf{v}^{(2)} + \mathbf{v}^{(3)}
$$

End displacements:

- Fixed member restrained: $\mathbf{v}^{(1)} = \mathbf{0}$
- Simply supported + temperature (free expansion): $\mathbf{v}^{(2)} = \mathbf{v}^{H}$

Compatibility:

$$
\mathbf{0} = \mathbf{v}^{H} + \mathbf{v}^{(3)}
\quad\Rightarrow\quad
\mathbf{v}^{(3)} = -\mathbf{v}^{H}
$$

### Computing Thermal Fixed-End Forces

Thus, the end forces in (3) that cancel thermal expansion in (2) are obtained from the local stiffness relation:

$$
\mathbf{q}^{H} = \mathbf{k} \,(-\mathbf{v}^{H})
$$

- $ \mathbf{q}^{H} $ — thermal fixed-end force vector
- $ \mathbf{v}^{H} $ — thermal end displacement vector  
- $ \mathbf{k} $ — local element stiffness matrix  


### Superposition Summary

To obtain the fixed-end forces for the restrained member:

1. Consider the simply supported member subjected to the temperature change
2. It develops the free end displacements $\mathbf{v}^{H}$ but no end forces
3. Apply end forces that produce the opposite displacement vector $-\mathbf{v}^{H}$

Those end forces are exactly the fixed-end forces required to suppress the free thermal deformation.


### Local FEFs from the Local Stiffness Matrix

$$
\mathbf{q}^H
=
\mathbf{k} \, (-\mathbf{v}^H)
$$


$$
{\scriptsize
=
\begin{bmatrix}
\dfrac{EA}{L} & 0 & 0 & -\dfrac{EA}{L} & 0 & 0 \\
0 & \dfrac{12EI}{L^3} & \dfrac{6EI}{L^2} & 0 & -\dfrac{12EI}{L^3} & \dfrac{6EI}{L^2} \\
0 & \dfrac{6EI}{L^2} & \dfrac{4EI}{L} & 0 & -\dfrac{6EI}{L^2} & \dfrac{2EI}{L} \\
-\dfrac{EA}{L} & 0 & 0 & \dfrac{EA}{L} & 0 & 0 \\
0 & -\dfrac{12EI}{L^3} & -\dfrac{6EI}{L^2} & 0 & \dfrac{12EI}{L^3} & -\dfrac{6EI}{L^2} \\
0 & \dfrac{6EI}{L^2} & \dfrac{2EI}{L} & 0 & -\dfrac{6EI}{L^2} & \dfrac{4EI}{L}
\end{bmatrix}
\left(
-\frac{\alpha L}{2}
\begin{bmatrix}
0 \\ 0 \\ -(T_b - T_t)/d \\ T_b + T_t \\ 0 \\ (T_b - T_t)/d
\end{bmatrix}
\right)
}
$$


### Local FEFs Vector

Assume the begin node is $b$ and the end node is $e$.

$$
\mathbf{q}^{H}
=
\begin{bmatrix}
q_{xb}^{H} \\[4pt]
q_{yb}^{H} \\[4pt]
q_{mb}^{H} \\[4pt]
q_{xe}^{H} \\[4pt]
q_{ye}^{H} \\[4pt]
q_{me}^{H}
\end{bmatrix}
=
E\alpha
\begin{bmatrix}
A(T_b + T_t)/2 \\[4pt]
0 \\[4pt]
I(T_b - T_t)/d \\[4pt]
-A(T_b + T_t)/2 \\[4pt]
0 \\[4pt]
-I(T_b - T_t)/d
\end{bmatrix}
$$


### Component Form

<div style="text-align:center;">
  <img src="assets/L2_Temp_FEF1.png" style="width:60%;">
</div>


The thermal fixed-end force expressions are therefore

$$
q_{xb}^{H} = -q_{xe}^{H} = EA\alpha\left(\frac{T_b + T_t}{2}\right)
$$

$$
q_{yb}^{H} = q_{ye}^{H} = 0
$$

$$
q_{mb}^{H} = -q_{me}^{H} = EI\alpha\left(\frac{T_b - T_t}{d}\right)
$$

### Special Case: Uniform Temperature Change

If the temperature increase is uniform through the depth,

$$
T_b = T_t = T_u
$$

then the gradient vanishes, so the bending terms disappear.

The fixed-end force vector becomes

$$
\mathbf{q}^{H}
=
\begin{bmatrix}
EA\alpha T_u \\[4pt]
0 \\[4pt]
0 \\[4pt]
-EA\alpha T_u \\[4pt]
0 \\[4pt]
0
\end{bmatrix}
$$

Uniform temperature change causes axial restraint force, but no bending moment.


### Beams, Frames, and Trusses

These expressions adapt naturally by member type:

- **Plane frame member**: both axial and bending terms may be present
- **Beam member**: axial expansion is free, so only the moment terms matter
- **Truss member**: member ends are free to rotate, so only the axial terms matter

In particular:

- For a **beam**: $q_{xb}^{H}=0$ and $q_{xe}^{H}=0$  
- For a **truss**: $q_{mb}^{H}=0$ and $q_{me}^{H}=0$

### Where Thermal FEFs Enter the DSM Equations

At the element level,

$$
\mathbf{q}_i = \mathbf{k}_{l,i}\mathbf{v}_i + \mathbf{q}_i^{F} + \mathbf{q}_i^{H}
$$

At the assembled global level,

$$
\mathbf{f} = \mathbf{K}\mathbf{u} + \mathbf{f}^{F} + \mathbf{f}^{H}
$$

or, equivalently,

$$
\mathbf{K}\mathbf{u} = \mathbf{f} - \mathbf{f}^{F} - \mathbf{f}^{H}
$$

So thermal effects enter the system exactly like other equivalent nodal force terms.


## Part 4 — Fabrication Errors Conceptual Framework

### What Is a Fabrication Error?

A fabrication error is a small initial deformation or geometric imperfection of a member in its unstressed state.

Common examples include:

- the member being fabricated too long or too short
- the member not being perfectly straight

If the structure must force that imperfect member into the intended geometry, internal forces develop.


### Same Superposition Logic as Temperature

The derivation is parallel to the thermal case:

1. Determine the member’s free end displacements caused by the geometric error
2. Apply equal and opposite end actions to suppress those displacements
3. The required end actions are the fabrication-error fixed-end forces

We will denote these by

$$
\mathbf{q}_i^{\varepsilon}
$$


### Two Common Types of Fabrication Error

In this lecture we consider:

1. **Error in initial member length**
2. **Error in initial member straightness**

These lead to different forms of $\mathbf{q}_i^{\varepsilon}$.

### Fabrication Error in Member Length

Suppose the design length is $L$, but the unstressed fabricated length is

$$
L + e_a
$$

with $e_a > 0$ meaning the member is too long.

To fit the member between supports separated by distance $L$, the structure must compress it axially.

That induces the local fabrication-error fixed-end force vector

$$
\mathbf{q}_i^{\varepsilon, a}
=
\begin{bmatrix}
EAe_a/L \\[4pt]
0 \\[4pt]
0 \\[4pt]
-EAe_a/L \\[4pt]
0 \\[4pt]
0
\end{bmatrix}
$$


### Component Form for Length Error

$$
N_b^{\varepsilon,a} = -N_e^{\varepsilon,a} = \frac{EA}{L}e_a
$$

$$
V_b^{\varepsilon,a} = V_e^{\varepsilon,a} = 0
$$

$$
M_b^{\varepsilon,a} = M_e^{\varepsilon,a} = 0
$$

This expression also applies directly to truss members, since the effect is purely axial.

### End Rotations Caused by the Initial Bend

Let the initial straightness error be $e_b$, occurring at a point that divides the member into lengths $l_1$ and $l_2$, with

$$
L = l_1 + l_2
$$

Then the corresponding end rotations are

$$
\theta_b = \frac{e_b}{l_1}
\qquad
\theta_e = \frac{e_b}{l_2}
$$

Using our local sign convention:

- the begin-end rotation is clockwise, so it enters as negative
- the end-node rotation is counterclockwise, so it enters as positive


### Local End Displacement Vector Due to Straightness Error

For a beam/frame member, the local end displacement vector due to the initial bend is

$$
\mathbf{v}_i^{\varepsilon,b}
=
\begin{bmatrix}
0 \\[4pt]
0 \\[4pt]
-\theta_b \\[4pt]
0 \\[4pt]
0 \\[4pt]
\theta_e
\end{bmatrix}
=
e_b
\begin{bmatrix}
0 \\[4pt]
0 \\[4pt]
-1/l_1 \\[4pt]
0 \\[4pt]
0 \\[4pt]
1/l_2
\end{bmatrix}
$$


### Local FEF Vector Due to Straightness Error

The corresponding local fixed-end force vector is

$$
\mathbf{q}_i^{\varepsilon,b}
=
\begin{bmatrix}
0 \\[6pt]
\dfrac{6EIe_b}{L^2 l_1 l_2}(l_2-l_1) \\[10pt]
\dfrac{2EIe_b}{L l_1 l_2}(2l_2-l_1) \\[10pt]
0 \\[6pt]
-\dfrac{6EIe_b}{L^2 l_1 l_2}(l_2-l_1) \\[10pt]
\dfrac{2EIe_b}{L l_1 l_2}(l_2-2l_1)
\end{bmatrix}
$$


### Component Form for Straightness Error

$$
N_b^{\varepsilon,b} = N_e^{\varepsilon,b} = 0
$$

$$
V_b^{\varepsilon,b} = -V_e^{\varepsilon,b} = \frac{6EIe_b}{L^2 l_1 l_2}(l_2-l_1)
$$

$$
M_b^{\varepsilon,b} = \frac{2EIe_b}{L l_1 l_2}(2l_2-l_1)
$$

$$
M_e^{\varepsilon,b} = \frac{2EIe_b}{L l_1 l_2}(l_2-2l_1)
$$

Since the straightness error is assumed small, it produces no axial deformation, only bending-type restraint effects.


### Total Fabrication-Error FEFs

If multiple fabrication-error sources are present, their local fixed-end force vectors add:

$$
\mathbf{q}_i^{\varepsilon}
=
\mathbf{q}_i^{\varepsilon,a}
+
\mathbf{q}_i^{\varepsilon,b}
+ \cdots
$$

So fabrication error is treated just like any other equivalent nodal load source.


## Part 6 — DSM Implementation

### Notation Summary

Local equivalent nodal force vectors:

$$
\mathbf{q}_i^{F}
\quad \text{member-load FEFs}
$$

$$
\mathbf{q}_i^{\theta}
\quad \text{temperature-change FEFs}
$$

$$
\mathbf{q}_i^{\varepsilon}
\quad \text{fabrication-error FEFs}
$$

Total local end force equation:

$$
\mathbf{q}_i = \mathbf{k}_{l,i}\mathbf{v}_i + \mathbf{q}_i^{F} + \mathbf{q}_i^{\theta} + \mathbf{q}_i^{\varepsilon}
$$


### Global Assembly Statement

After transformation and assembly,

$$
\mathbf{f} = \mathbf{K}\mathbf{u} + \mathbf{f}^{F} + \mathbf{f}^{\theta} + \mathbf{f}^{\varepsilon}
$$

or, in solved form,

$$
\mathbf{K}\mathbf{u} = \mathbf{f} - \mathbf{f}^{F} - \mathbf{f}^{\theta} - \mathbf{f}^{\varepsilon}
$$

So the implementation pattern is the same one we have already used throughout DSM.


### Implementation Steps

1. Build each element local stiffness matrix $\mathbf{k}_{l,i}$
2. Compute any element-level vectors $\mathbf{q}_i^{F}$, $\mathbf{q}_i^{\theta}$, $\mathbf{q}_i^{\varepsilon}$
3. Transform to global coordinates
4. Assemble the global equivalent nodal force vectors
5. Solve the global system

$$
\mathbf{K}\mathbf{u} = \mathbf{f} - \mathbf{f}^{F} - \mathbf{f}^{\theta} - \mathbf{f}^{\varepsilon}
$$

6. Perform the backward pass to recover local element end forces

## Wrap-Up

Temperature changes and fabrication errors are handled in DSM as **equivalent nodal force effects**.

For temperature changes:

- average temperature produces axial restraint force
- temperature gradient produces end moments

For fabrication errors:

- length error produces axial restraint force
- straightness error produces shear/moment restraint effects

At the element level:

$$
\mathbf{q}_i = \mathbf{k}_{l,i}\mathbf{v}_i + \mathbf{q}_i^{F} + \mathbf{q}_i^{\theta} + \mathbf{q}_i^{\varepsilon}
$$

At the global level:

$$
\mathbf{K}\mathbf{u} = \mathbf{f} - \mathbf{f}^{F} - \mathbf{f}^{\theta} - \mathbf{f}^{\varepsilon}
$$
